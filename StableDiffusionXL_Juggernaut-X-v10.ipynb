```python
from diffusers import DiffusionPipeline
import torch

# see https://huggingface.co/RunDiffusion/Juggernaut-X-v10

pipe = DiffusionPipeline.from_pretrained(
    'RunDiffusion/Juggernaut-X-v10',
    torch_dtype = torch.float16)

pipe.to('cuda');
```

```python
# see https://rundiffusion.com/juggernaut-xl#prompting-guide

prompt = 'A dreamy and fantastical photo of a unicorn riding a hoverboard, with a backdrop of the famous Golden Gate Bridge in San Francisco'
negative_prompt = 'cartoon, deformed, bad composition, cgi, 3D, digital, airbrushed, hands, hand'

image = pipe(prompt = prompt, negative_prompt = negative_prompt).images[0]
display(image)
```

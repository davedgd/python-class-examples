```python
from IPython.display import Audio
import os
import torch
import torchaudio
from seamless_communication.models.inference import Translator
```

```python
audio_file = 'https://github.com/neonbjb/tortoise-tts/raw/main/tortoise/voices/pat/1.wav'

import os
os.system('wget ' + audio_file + ' -nc')

display(Audio(audio_file, autoplay = False))
```

```python
translator = Translator('seamlessM4T_medium', 'vocoder_36langs', torch.device('cuda:0'))
```

```python
tgt_lang = 'fra' # for supported languages, see: https://github.com/facebookresearch/seamless_communication/tree/main/scripts/m4t/predict#supported-languages
output_file = 'out.wav'

translated_text, wav, sr = translator.predict('1.wav', 's2st', tgt_lang)

torchaudio.save(
    output_file,
    wav[0].cpu(),
    sample_rate = sr
)

display(Audio(output_file, autoplay = False))
```

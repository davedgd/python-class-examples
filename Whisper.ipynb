```python
from IPython.display import Audio
import whisper
```

```python
audio_to_transcribe = 'https://github.com/neonbjb/tortoise-tts/raw/main/tortoise/voices/pat/1.wav'

display(Audio(audio_to_transcribe, autoplay = False))
```

```python
model = whisper.load_model('base') # choose from tiny, base, small, medium, large (v2) 
result = model.transcribe(audio_to_transcribe)
print(result["text"])
```

```python
import torch
import torchaudio
from IPython.display import Audio
from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_voice
from transformers import logging

logging.set_verbosity_error()

tts = TextToSpeech(use_deepspeed = False, 
                   kv_cache = True, 
                   half = True)
```

```python
# see built-in voice options (you can create your own)
%ls tortoise/voices
Audio('tortoise/voices/tom/1.wav')
```

```python
# settings
voice = 'tom'
preset = 'ultra_fast' # quality setting -- choose from: 'ultra_fast', 'fast', 'standard', or 'high_quality'
text = "Thanks so much for coming to this M B A workshop! I hope you all learned a lot about what's been going on lately in the AI space."
```

```python
voice_samples, conditioning_latents = load_voice(voice)
gen = tts.tts_with_preset(text, 
                          voice_samples = voice_samples, 
                          conditioning_latents = conditioning_latents, 
                          use_deterministic_seed = 5473, 
                          preset = preset)
torchaudio.save('../generated.wav', gen.squeeze(0).cpu(), 24000)
Audio('../generated.wav')
```
